```python
import torch
```

```python
class Neuron:
  def __init__(self, n_features):
    self.w = torch.rand(n_features, 1, requires_grad=True)
    self.b = torch.rand(1,1, requires_grad=True)

  def __call__(self, x):
    z = self.w.T @ x + self.b
    a = torch.tanh(z)
    return a

  def parameters(self):
    return [self.w , self.b]

class Layer:
  def __init__(self, n_in, n_out):
    self.neurons = [Neuron(n_in) for _ in range(n_out)]

  def __call__(self, X):
    outs = [neuron(X) for neuron in self.neurons]
    return torch.cat(outs, dim=0)

  def parameters(self):
      return [param for neuron in self.neurons for param in neuron.parameters()]
  
class MLP:
  def __init__(self, n_in, n_outs):
    n_input = [n_in] + n_outs
    self.layers = [Layer(n_input[i], n_input[i+1]) for i in range(len(n_outs))]
  
  def __call__(self, x):
      for layer in self.layers:
        x = layer(x)
      return x
  
  def parameters(self):
        return [param for layer in self.layers for param in layer.parameters()]
```

```python
X = torch.rand(4,3)
y = torch.rand(4,1)
mlp = MLP(X.shape[1], [3,3,1])
```

```python
for _ in range(100):
  y_pred = torch.cat([mlp(x.unsqueeze(1)) for x in X], dim=0)

  loss = ((y_pred - y) ** 2).mean()

  loss.backward()

  with torch.no_grad():
    for param in mlp.parameters():
      param -= 0.01 * param.grad

  for param in mlp.parameters():
       param.grad.zero_()

  print(loss.item())
```
